```python
# salva imagens aumentadas na pasta output_dir,rotina de augmentação separada

dataset_dir = 'dataset/dataset'
output_dir = 'dataset/dataset_augmented'

augmenter = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05,
    zoom_range=0.05,
    brightness_range=[0.9, 1.1],
    horizontal_flip=True
)

for class_name in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_name)
    save_path = os.path.join(output_dir, class_name)
    os.makedirs(save_path, exist_ok=True)

    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        img = load_img(img_path)
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)  # (1, altura, largura, canais)

        i = 0
        for batch in augmenter.flow(x, batch_size=1, save_to_dir=save_path, save_prefix='aug', save_format='jpg'):
            i += 1
            if i >= 5:  # número de imagens aumentadas que quer salvar por original
                break
```

```python

# 1. Verificar caminho dataset_dir e criar pasta de saída
dataset_dir = 'dataset/dataset'
if not os.path.exists(dataset_dir):
    raise FileNotFoundError(f"Diretório do dataset não encontrado: {dataset_dir}")
if not os.path.isdir(dataset_dir):
    raise NotADirectoryError(f"O caminho do dataset não é uma pasta: {dataset_dir}")

print(f"Dataset localizado em: {dataset_dir}")

output_dir = os.path.join(os.path.dirname(dataset_dir), 'dataset_augmented')
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Pasta para imagens aumentadas criada em: {output_dir}")
else:
    print(f"Pasta para imagens aumentadas já existe: {output_dir}")

# 2. Agora configure o ImageDataGenerator para usar dataset_dir
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05,
    zoom_range=0.05,
    brightness_range=[0.9, 1.1],
    horizontal_flip=True
)

train_generator = train_datagen.flow_from_directory(
    dataset_dir,
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',
    subset='training'
)
```

```python
# Atribuir labels automaticamente com ImageDataGenerator com aumento de dados
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05,
    zoom_range=0.05,
    brightness_range=[0.9, 1.1],
    horizontal_flip=True
)

train_generator = train_datagen.flow_from_directory(
    'dataset/dataset',
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',
    subset='training'
)

class_indices = train_generator.class_indices
idx_to_name = {v: k for k, v in class_indices.items()}


val_generator = train_datagen.flow_from_directory(
    'dataset/dataset',
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',
    subset='validation'
)
```

```python
# Visualizar imagens do train_generator

# Pega um lote de imagens
images, labels = next(train_generator)

# Mostra as primeiras 8 imagens com o nome da classe
class_names = list(train_generator.class_indices.keys())

plt.figure(figsize=(15, 5))
for i in range(8):
    plt.subplot(2, 4, i + 1)
    plt.imshow(images[i])
    label_idx = np.argmax(labels[i])
    plt.title(class_names[label_idx])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Treinar um modelo com MobileNetV2 (Transfer Learning)
# Carregar a MobileNetV2 pré-treinada
# Usar como base para classificar os rostos
# Treinar com seu dataset

# Carregar a MobileNetV2 sem o topo (camadas de classificação)
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Congelar pesos da base (não treinar essas camadas)
base_model.trainable = False

# Adicionar camadas personalizadas
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
predictions = Dense(train_generator.num_classes, activation='softmax')(x)

# Montar o modelo final
model = Model(inputs=base_model.input, outputs=predictions)

# Compilar
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Treinar
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=100,
    callbacks=[EarlyStopping(patience=5, restore_best_weights=True)],
    verbose=1
)
```

```python
# Criar e compilar o modelo com base congelada
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
predictions = Dense(train_generator.num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Primeiro treinamento com base congelada
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=100,
    callbacks=[EarlyStopping(patience=5, restore_best_weights=True)],
    verbose=1
)

# AGORA ENTRA AQUI: Fine-tuning (após o primeiro fit)
# Descongela as últimas 30 camadas
base_model.trainable = True
for layer in base_model.layers[:-30]:
    layer.trainable = False

# Recompila com uma taxa de aprendizado menor (ajuste fino)
model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 4. Novo treinamento com base parcialmente treinável
history_finetune = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,
    callbacks=[EarlyStopping(patience=3, restore_best_weights=True)],
    verbose=1
)
```

```python
# Definir margem para ampliar a caixa do rosto

# Supõe que você já tenha:
# - 'image_rgb' (PIL.Image)
# - 'results' (rostos detectados com MTCNN)
# - 'model' (modelo treinado)
# - 'idx_to_name' (dicionário de índices para nomes)

draw = ImageDraw.Draw(image_rgb)
width_img, height_img = image_rgb.size
margin = 0.4

predicted_classes = []  # ← Lista para armazenar as classes previstas

for face in results:
    x, y, width, height = face['box']

    # Calcular região com margem
    x1 = int(max(x - width * margin, 0))
    y1 = int(max(y - height * margin, 0))
    x2 = int(min(x + width * (1 + margin), width_img))
    y2 = int(min(y + height * (1 + margin), height_img))

    # Recortar e preprocessar o rosto para o modelo
    face_crop = image_rgb.crop((x1, y1, x2, y2))
    face_crop = face_crop.resize((224, 224))  # Tamanho esperado pelo modelo
    face_array = np.array(face_crop) / 255.0  # Normalizar
    face_array = np.expand_dims(face_array, axis=0)  # Adicionar batch

    # Prever
    prediction = model.predict(face_array)
    class_idx = np.argmax(prediction)
    predicted_classes.append(class_idx)

# Agora que temos as previsões, desenhar tudo
for i, face in enumerate(results):
    x, y, width, height = face['box']
    x1 = int(max(x - width * margin, 0))
    y1 = int(max(y - height * margin, 0))
    x2 = int(min(x + width * (1 + margin), width_img))
    y2 = int(min(y + height * (1 + margin), height_img))

    draw.rectangle(((x1, y1), (x2, y2)), outline=(0, 255, 0), width=3)
    name = idx_to_name[predicted_classes[i]]
    draw.text((x1, y1 - 20), name, fill=(255, 0, 0))



```
