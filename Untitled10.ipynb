```python
# Inicializar detectores e modelos
detector = MTCNN()
embedder = FaceNet()
```

```python
# Função para extrair uma face da imagem
def extract_face(img_path, required_size=(160,160)):
    """
    Detecta a primeira face na imagem e recorta para o tamanho requerido.
    Retorna a face como numpy array ou None se não detectar.
    """
    img = Image.open(img_path).convert('RGB')
    pixels = np.asarray(img)
    results = detector.detect_faces(pixels)
    if not results:
        return None
    x, y, w, h = results[0]['box']
    x, y = abs(x), abs(y)
    face = pixels[y:y+h, x:x+w]
    face = Image.fromarray(face).resize(required_size)
    return np.asarray(face)
```

```python
# Função para extrair embedding da face
def get_embedding(face_pixels):
    """
    Recebe uma face (numpy array), normaliza e retorna o embedding (vetor 512D).
    """
    face_pixels = face_pixels.astype('float32')
    face_pixels /= 255.0
    samples = np.expand_dims(face_pixels, axis=0)
    embedding = embedder.model.predict(samples)
    return embedding[0]
```

```python
# Função para carregar dataset e extrair embeddings e labels
def load_dataset(dir_path):
    """
    Varre o dataset organizado por pastas com nomes das pessoas,
    extrai as faces, calcula embeddings e retorna arrays X e y.
    """
    X, y = [], []
    for person in os.listdir(dir_path):
        person_dir = os.path.join(dir_path, person)
        if not os.path.isdir(person_dir):
            continue
        for fname in os.listdir(person_dir):
            path = os.path.join(person_dir, fname)
            if os.path.isdir(path):
                continue
            face = extract_face(path)
            if face is None:
                print(f"[Aviso] Nenhuma face detectada em {path}")
                continue
            emb = get_embedding(face)
            X.append(emb)
            y.append(person)
        print(f"Carregadas imagens de: {person}, total embeddings até agora: {len(X)}")
    return np.asarray(X), np.asarray(y)
```

```python
# Carregar dataset e extrair embeddings
print("\nCarregando dataset e extraindo embeddings, aguarde...")
trainX, trainy = load_dataset('dataset/dataset')
```

```python
# Verificar dados carregados
if len(trainX) == 0 or len(trainy) == 0:
    raise ValueError("Dados de treino vazios. Verifique o dataset.")

print(f"Número de embeddings: {trainX.shape[0]}, Dimensão: {trainX.shape[1]}")
print(f"Pessoas (classes) no dataset: {np.unique(trainy)}")
```

```python
# Treinar classificador KNN
knn = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
knn.fit(trainX, trainy)
print("KNN treinado com sucesso!")
```

```python
# Salvar modelo para uso futuro
joblib.dump(knn, 'knn_face_classifier.joblib')
print("Modelo KNN salvo em 'knn_face_classifier.joblib'.")
```

```python
# Avaliação do modelo (usando dados de treino como teste, só para checagem)
y_pred = knn.predict(trainX)
print("\nRelatório de Classificação no Dataset de Treino:")
print(classification_report(trainy, y_pred))
```

```python
# Visualização dos embeddings com PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(trainX)

plt.figure(figsize=(8,6))
for person in np.unique(trainy):
    ix = np.where(trainy == person)
    plt.scatter(X_pca[ix, 0], X_pca[ix, 1], label=person)
plt.title("Visualização dos embeddings com PCA")
plt.legend()
plt.show()
```

```python
# Visualização dos embeddings com t-SNE
tsne = TSNE(n_components=2, random_state=42, init='pca')
X_tsne = tsne.fit_transform(trainX)

plt.figure(figsize=(8,6))
for person in np.unique(trainy):
    ix = np.where(trainy == person)
    plt.scatter(X_tsne[ix, 0], X_tsne[ix, 1], label=person)
plt.title("Visualização dos embeddings com t-SNE")
plt.legend()
plt.show()
```

```python
# Reconhecimento facial na imagem de grupo
group_image_path = "beatles_group.PNG"

print(f"Detectando faces na imagem de grupo: {group_image_path}")
img_group = Image.open(group_image_path).convert('RGB')
pixels = np.asarray(img_group)
results = detector.detect_faces(pixels)

print(f"Faces detectadas: {len(results)}")
draw = ImageDraw.Draw(img_group)
margin = 0.4
width_img, height_img = img_group.size

for i, face in enumerate(results):
    x, y, w, h = face['box']
    x, y = abs(x), abs(y)

    # Ajustar caixa com margem
    x1 = int(max(x - w * margin, 0))
    y1 = int(max(y - h * margin, 0))
    x2 = int(min(x + w * (1 + margin), width_img))
    y2 = int(min(y + h * (1 + margin), height_img))

    # Extrair face e gerar embedding
    face_crop = img_group.crop((x1, y1, x2, y2)).resize((160,160))
    face_array = np.asarray(face_crop)
    emb = get_embedding(face_array)

    # Prever com KNN
    pred_class = knn.predict([emb])[0]

    # Desenhar retângulo e nome
    draw.rectangle(((x1, y1), (x2, y2)), outline=(0, 255, 0), width=3)
    draw.text((x1, y1 - 20), pred_class, fill=(255, 0, 0))

# Mostrar imagem final com faces reconhecidas
plt.figure(figsize=(12,12))
plt.imshow(img_group)
plt.axis('off')
plt.title("Reconhecimento Facial: FaceNet + KNN")
plt.show()
```
